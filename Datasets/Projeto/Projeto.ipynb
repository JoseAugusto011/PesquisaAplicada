### Módulos

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

```

```python
def ValoresInvalidos(df):
    
   
    
    
    #Ver quais os valores únicos de cada coluna
    #Vericar qtd de repetições das variaveis em cada coluna 
    print("\n")
    print(f'Há duplicatas? {df.duplicated().values.any()}')
    print(f'Há valores NaN? {df.isna().values.any()}')
    print(f"Há valores nulos? {df.isnull().values.any()}")
    print("\n")
    
```

```python
def VerificarValores(df):
     
    print("Informações sobre os dados: \n")    
    print("Tamanho do dataset: ", df.shape) 
    #Mostrar os 10 valores com mais repetições no dicionario
    print("Valores mais repetidos: \n")
    for col in df.columns:
        print(f'Coluna: {col}')
        print(df[col].value_counts().head(3)) 
        print("\n\n")
        
        
    
```

```python
def RemoveColunas(df):
    df.drop([df.columns[4],df.columns[5],df.columns[6],df.columns[7],df.columns[8],df.columns[9],df.columns[13]], axis=1, inplace=True)
    return df
```

### Análise primária

```python
ap1 = pd.read_csv('AntesPandemiaI.csv', sep=';')
ap2 = pd.read_csv('AntesPandemiaII.csv', sep=';')
ap3 = pd.read_csv('AntesPandemiaIII.csv', sep=';')
ap4 = pd.read_csv('AntesPandemiaIV.csv', sep=';')

dp = pd.read_csv('DurantePandemia.csv', sep=';')
pp = pd.read_csv('PosPandemia.csv', sep=';')
```

```python
print("Antes da Pandemia I")
ValoresInvalidos(ap1)
VerificarValores(ap1)
```

```python
print("Antes da Pandemia II")
ValoresInvalidos(ap2)
VerificarValores(ap2)
```

```python
print("Antes da Pandemia III")
ValoresInvalidos(ap3)
VerificarValores(ap3)
```

```python
print("Antes da Pandemia IV")
ValoresInvalidos(ap4)
VerificarValores(ap4)
```

```python
print("Durante a Pandemia")
ValoresInvalidos(dp)
VerificarValores(dp)
```

```python
print("Depois da Pandemia")
ValoresInvalidos(pp)
VerificarValores(pp)
```
